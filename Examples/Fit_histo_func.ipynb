```python
import numpy as np
import pandas as pd
import sys 
sys.path.append('/home/riccardo/GraRed')
from GraRed import Catalogue
import matplotlib.pyplot as plt
from GraRed import utils as ut
import itertools
from GraRed import Fit_bckg_corr as fbc
from GraRed import plot_utils as pu
```

```python
df =pd.read_parquet('/home/riccardo/GraRed/Data/SDSS_galaxy_cat.parquet')
gal = Catalogue.Galaxies(df['RA[deg]'].values,df['DEC[deg]'].values,df['Z'].values,redshift_err=df['zerr'].values)

df =pd.read_parquet('/home/riccardo/GraRed/Data/WH15_cluster_cat.parquet')
m = 1.08*np.log10(df.RL500.values) - 1.37 +14
df['M500']=10**m

rand_gen = np.random.default_rng(12345)


dd = df.loc[rand_gen.choice(df.index,10000,replace=False)]

cl = Catalogue.Cluster(dd['RA[deg]'].values,dd['Dec[deg]'].values,dd['z'].values,mass=dd['M500'].values,radius=dd['R500[Mpc]'].values,
concentration=None,is_crit=True, delta=500,gravity='GR', seed=1234)

dd=pd.read_parquet('/home/riccardo/catalogo.parquet')
```

```python
df=dd[dd.num_mean_centr > 1]
r=df.dist_member
rarray=list(itertools.chain.from_iterable(r))
varray=list(itertools.chain.from_iterable(df.vel_member/1000))
histo,xedges,yedges=fbc.create_histo(rarray,varray,0.05,0.05)
plt.figure()
plt.imshow(histo.T,vmin=0.001,vmax=30,cmap=pu.my_map("viridis",2408,"white","yellow"),extent=[0,4,-4,4])
plt.colorbar()
plt.title("Histogram_2D of distances and velocities of galaxies")
plt.xlabel("r/r500")
plt.ylabel("v(10^3 km/s)")
plt.show()
```

```python
fit_func,par=fbc.fit_histo(histo,xedges,yedges,3,"Lin_2")
```

```python
new_hist=fbc.hist_bckg_rem(histo,fit_func)
plt.figure()
plt.imshow(new_hist,vmin=0.001,vmax=30,cmap=pu.my_map("viridis",2408,"white","yellow"),extent=[0,4,-4,4])
plt.colorbar()
plt.title("Histogram_2D of distances and velocities of galaxies after background removal")
plt.xlabel("r/r500")
plt.ylabel("v(10^3 km/s)")
plt.show()
```

```python
plt.figure()
plt.hist(histo.ravel(),range=[np.min(histo),10],label="hist")
plt.hist(new_hist.ravel(),range=[np.min(new_hist),10],label="new_hist")
plt.legend()
plt.title("Histogram_2D of occurences")
plt.xlabel("counts")
plt.ylabel("counts of counts")
plt.show()
```

```python
histog=fbc.histo_cut_norm(4,histo,fit_func,xedges,yedges)
offset=[-0.2,0.3,0.6,0.9]

plt.figure()
for i in range(len(histog[:,0])):
    plt.scatter(yedges,histog[i,]+offset[i],label=str(i)+"<r<"+str(i+1),s=1)
plt.legend()
plt.ylim(-0.25,1.7)
plt.show()
```

```python
err=fbc.err_norm(4,histo,fit_func,xedges,yedges,3)

plt.figure()
for i in range(len(histog[:,0])):
    plt.errorbar(yedges,histog[i,]+offset[i],yerr=err[i,:],label=str(i)+"<r<"+str(i+1),fmt="o",markersize=1)
plt.legend()
plt.ylim(-0.25,1.7)
plt.title("fit of velocity distribution")
plt.xlabel("v(10^3 km/s)")
plt.ylabel("Pdf")
plt.show()
```
